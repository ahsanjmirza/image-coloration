```python
# IMPORT DATA & MODEL
from wgan_gp import WGAN_GP
from dataset import VintageFacesDataset

### TRAIN
train_data = VintageFacesDataset(
    img_dir                     = './UTKFaces/train', 
    size                        = 256
)

wgan_gp = WGAN_GP(
    in_channels_gen             = 1,
    init_features_gen           = 32,
    out_channels_gen            = 3,
    in_channels_disc            = 3,
    init_features_disc          = 8
)

wgan_gp.train(
    train_data                  = train_data,
    epochs                      = 55,
    steps_per_epoch             = 1500,
    batch_size                  = 4,
    lr_gen                      = 1e-4,
    betas_gen                   = (.5, .99),
    lr_disc                     = 1e-4,
    betas_disc                  = (.5, .99),
    disc_iter                   = 5,
    verbose_every               = 50,
    save_every_epoch            = True,
    save_weights_dir            = './model_weights',
    gp_lambda                   = 10,
    loss_gamma                  = 0.8,
    continue_epoch              = 2
)

del wgan_gp, train_data
```

```python
# IMPORT DATA & MODEL
from wgan_gp import WGAN_GP
from dataset import VintageFacesDataset

### TEST
test_data = VintageFacesDataset(
    img_dir                     = './test/original', 
    size                        = 256
)

wgan_gp = WGAN_GP(
    in_channels_gen             = 1,
    init_features_gen           = 32,
    out_channels_gen            = 3,
    in_channels_disc            = 3,
    init_features_disc          = 8
)

wgan_gp.load_gen('./model_weights/generator/Epoch_2.pth')
wgan_gp.test(test_data, './test/reconstructed')

del wgan_gp, test_data
```
